# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python

#generate a gaussian r.v X
x = stats.norm.rvs(size=1000);
#here we are creating 1000 random number following an normal distribution
sns.set_style('whitegrid')
#setting the grid style as whitefrid
sns.kdeplot(np.array(x), bw=0.7,shade=True)
#here we are plotting the random numbe stored in x, using an kernel density method, where  the bandwidth of kernet is 0.5
#here we are also shading the area under the kernel
#sns.kdeplot()
sns.kdeplot(np.array(x), bw=0.7,shade=True,cumulative=True)
#for displaying the CDF of the distribution x
plt.show()
#here we are displaying the plot
```

```python
stats.kstest(x, 'norm')
#here we are performing the KS test on X and comparing it with the normal distribution
```

```python
y = np.random.uniform(0,1,100000);
# Y ~ Continous Uniform Distribution(0,1) containing 10000 points
sns.kdeplot(np.array(y), bw=0.1,shade=True)
#here we first converting the dataset to an 1d array as it is the default input to kdeplot, then using an bw to estimate the pdf
sns.kdeplot(np.array(y), bw=0.1,shade=True,cumulative=True)
#here we are plotting the CDF of dataset in the same plot with same configuration as above
plt.show()
#here we are displaying the plot 
```

```python
stats.kstest(y, 'uniform')
```

#So from the above result we can see clearly that our generated distribution and output from KS test are matching
