# Empirical bootstrap based Confidence Interval

```python
import numpy
from pandas import read_csv
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
from matplotlib import pyplot
import matplotlib.pyplot as plt

import seaborn as sns
from scipy import stats
```

```python

# load dataset
x = numpy.array([180,162,158,172,168,150,171,183,165,176])
print(numpy.median(x))
# configure bootstrap
n_iterations = 1000
#number of samples to be generated by using bootstrap
n_size = 13
#sample size for boot strap
# run bootstrap
medians = list()
#creating a list of median for storing the medians
for i in range(n_iterations):
    # prepare train and test sets
    s = resample(x, n_samples=n_size);
    #here we are generating random sample from without replacement from resampler function using data X
    m = numpy.median(s);
    #calculating the median of the generated sample
    #print(m)
    medians.append(m)
    #here we are appending the median of the generated sample and appending it to the list of medians

sns.set_style('whitegrid')
sns.kdeplot(numpy.array(medians),bw=0.8,shade=True)
#plotting the KDE plot of the given medians, with bandwidth =1
#sns.kdeplot(numpy.array(medians),bw=1,shade=True,cumulative=True)
plt.show()
# plot scores
pyplot.hist(medians)
#here we are plotting the histogram of the medians
pyplot.show()
#displaying the plot using pyplot

# confidence intervals
alpha = 0.95
#here we want to find the CI with confidence of  95 %
p = ((1.0-alpha)/2.0) * 100
lower =  numpy.percentile(medians, p)
#finding the percentile value of lower median

p = (alpha+((1.0-alpha)/2.0)) * 100
upper =  numpy.percentile(medians, p)
#finding the percentile value of upper median
print('%.1f confidence interval %.1f and %.1f' % (alpha*100, lower, upper))
#printing the values as obtained from the calculations
```
