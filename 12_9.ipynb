# Q-Q Plot

```python
import numpy as np
import scipy.stats as stats
import pylab
```

```python
std_norm=np.random.normal(loc=5,scale=15,size=10000)
for i  in range(1,101):
  print(i,np.percentile(std_norm,i))
```

```python
stats.probplot(std_norm,dist='norm',plot=pylab)
pylab.show()
```

```python
std_uniform=np.random.uniform(low=-3,high=3,size=10000)
for i  in range(1,101):
  print(i,np.percentile(std_uniform,i))
```

```python
stats.probplot(std_uniform,dist='uniform',plot=pylab)
pylab.show()
```

```python
stats.probplot(std_uniform,dist='norm',plot=pylab)
pylab.show()
```
